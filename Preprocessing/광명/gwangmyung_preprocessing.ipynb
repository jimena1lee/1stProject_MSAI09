### 2. 광명시 인구및세대 현황
Raw-Data\광명_202512_202512_연령별인구현황_연간.csv\
출처: https://jumin.mois.go.kr/ageStatMonth.do\
행정구역: 경기도 광명시
기간: 연간 / 2025년 - 2025년\
연령구분단위: 5세\
연령조회범위: 0세 ~ 100세

```python
import pandas as pd
population_df = pd.read_csv(r'C:\Users\EL36\Desktop\1차프로젝트_CCTV\1stProject_MSAI09\Raw-Data\광명_202512_202512_연령별인구현황_연간.csv', encoding='cp949')
population_df.info()
population_df.head(3)
```

```python
# dtype = object인 모든 숫자열을 숫자형으로 변환

id_cols = ['행정구역']
num_cols = population_df.columns.difference(id_cols)

population_df[num_cols] = (
    population_df[num_cols]
    .replace(',', '', regex=True)
    .apply(pd.to_numeric)
)
```

```python
# 행정구역에서 'ㅇㅇㅇ동' 발라내기
population_df['동'] = population_df['행정구역'].str.extract(r'(\S+동)')

population_df = population_df[['동','2025년_계_총인구수','2025년_계_0~4세','2025년_계_5~9세','2025년_계_10~14세']]
population_df.head(3)
```

```python
# '동' 열에 결측치가 있는 행 제거   
population_df = population_df.dropna(subset=['동'])

# 컬럼 이름 변경
population_df = population_df.rename(columns={
    '2025년_계_총인구수': '총인구수',
    '2025년_계_0~4세': '0~4세',
    '2025년_계_5~9세': '5~9세',
    '2025년_계_10~14세': '10~14세'
})

population_df
```

```python
#어린이 총인구 컬럼 추가
population_df['어린이 총인구'] = population_df['0~4세'] + population_df['5~9세'] + population_df['10~14세']
```

```python
# 광명1동~광명7동, 철산1동~철산4동, 하안1동~하안3동, 소하1동~소하2동 을 광명동, 철산동, 하안동, 소하동으로 통합
import re

population_df['행정동_그룹'] = population_df['동'].str.replace(r'\d+동$', '동', regex=True)

#행정동별로 합산
agg_df = (
    population_df
        .groupby('행정동_그룹', as_index=False)
        .agg({
            '총인구수': 'sum',
            '0~4세': 'sum',
            '5~9세': 'sum',
            '10~14세': 'sum',
            '어린이 총인구': 'sum'
        })
)

#어린이 비율 계산

agg_df['어린이 비율(%)'] = (
    agg_df['어린이 총인구'] / agg_df['총인구수'] * 100
)

#정렬
agg_df = agg_df.sort_values('행정동_그룹').reset_index(drop=True)
```

```python
agg_df = agg_df.rename(columns={'행정동_그룹': '동'})
```

### 3. 광명시 사고데이터 정리

```python
acc = acc[acc['시군명'].str.contains('광명')]
acc = acc[['시군명', '구/읍면동명', '사고년도','발생건수', '사상자수', '사망자수', '중상자수', '경상자수', '부상자수', '위도', '경도']]
acc.head(3)
```

```python
#전국교통사고다발지역표준데이터 전처리
acc2 = pd.read_csv(r'C:\Users\EL36\Desktop\1차프로젝트_CCTV\1stProject_MSAI09\Raw-Data\스쿨존 사고데이터\전국교통사고다발지역표준데이터.csv', encoding='cp949')
acc2.columns
```

```python
acc2 = acc2[acc2['사고다발지역시도시군구'].str.contains('광명')]
acc2 = acc2[acc2['사고유형구분'].str.contains('어린이')]
acc2['구/읍면동명'] = acc2['사고지역위치명'].str.extract(r'\s([^\s()]+동)\(')
acc2['보호구역명'] = acc2['사고지역위치명'].str.extract(r'\(([^)]+ )\부근')
acc2 = acc2[['사고다발지역시도시군구','구/읍면동명','보호구역명','사고연도','사고건수','사상자수', '사망자수', '중상자수', '경상자수', '부상신고자수', '위도', '경도']]
acc2 = acc2.rename(columns={'사고다발지역시도시군구': '시군명', '사고연도': '사고년도', '사고건수': '발생건수', '부상신고자수': '부상자수'})
acc2.head(3)
```

```python
#스쿨존 내 사고 다발지와 전국교통사고다발지역표준데이터의 사고 다발지 합치기    
acc_combined = pd.concat([acc, acc2.drop(columns=['보호구역명'])], ignore_index=True)
#위도, 경도, 사고년도 기준으로 중복 제거
acc_combined = acc_combined.drop_duplicates(subset=['위도', '경도', '사고년도']).reset_index(drop=True)
acc_combined.head(3)
```

```python
# 위도 경도에 맞춰 인근 어린이보호구역 매칭하기

# 필요 함수
# 1. 두 좌표 간 거리 계산 함수
import glob
import os

def haversine_m(lat1, lon1, lat2, lon2):
    """
    두 좌표 사이의 거리를 미터(m) 단위로 반환.
    지구를 구체로 가정하는 Haversine 공식 사용.
    """
    R = 6371000  # 지구 반경 (미터)
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    return R * 2 * np.arcsin(np.sqrt(a))

acc_coordinates = acc_combined[['위도', '경도']].drop_duplicates()
school_coordinates = school_df[['대상시설명','위도', '경도']].drop_duplicates()

matched_zones = []
for _, acc_row in acc_coordinates.iterrows():
    for _, school_row in school_coordinates.iterrows():
        distance = haversine_m(acc_row['위도'], acc_row['경도'], school_row['위도'], school_row['경도'])
        if distance <= 300:  # 300m 이내인 경우
            matched_zones.append({
                '사고위도': acc_row['위도'],
                '사고경도': acc_row['경도'],
                '보호구역명': school_row['대상시설명']
            })

matched_df = pd.DataFrame(matched_zones)
matched_df.head(3)
```

```python
#사고위도와 사고경도를 기준으로 보호구역명 매칭 결과를 acc_combined에 병합
acc_combined = acc_combined.merge(matched_df, left_on=['위도', '경도'], right_on=['사고위도', '사고경도'], how='left')
acc_combined = acc_combined.drop(columns=['사고위도', '사고경도'])
acc_combined.head(3)
```

```python
acc_combined.info()
acc_combined[acc_combined.isna().any(axis=1)]
```

```python
acc_combined
```

```python
# 결측치 처리 - 보호구역명 결측치는 '보호구역명 없음'으로 대체
acc_combined['보호구역명'] = acc_combined['보호구역명'].fillna('보호구역명 없음')

# 2020~2024년 사고 데이터만 남기기
acc_combined = acc_combined[acc_combined['사고년도'] >= 2020]
acc_combined.info()
```

### 5. 안전시설물 위치와 보호구역 매칭

```python
import pandas as pd
import glob
import os

# 두 좌표 간 거리 계산 함수 
def haversine_m(lat1, lon1, lat2, lon2):
    """
    두 좌표 사이의 거리를 미터(m) 단위로 반환.
    지구를 구체로 가정하는 Haversine 공식 사용.
    """
    R = 6371000  # 지구 반경 (미터)
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    return R * 2 * np.arcsin(np.sqrt(a))
```

```python
base_path = r'C:\Users\EL36\Desktop\1차프로젝트_CCTV\1stProject_MSAI09\Preprocessing\광명'
cctv = pd.read_csv(os.path.join(base_path, '2_gwangmyung_cctv.csv'), encoding='cp949')
crosswalk = pd.read_csv(os.path.join(base_path, '2_gwangmyung_crosswalk.csv'), encoding='cp949')
yellow_carpet = pd.read_csv(os.path.join(base_path, '2_gwangmyung_yellow_carpet.csv'), encoding='cp949')
signal = pd.read_csv(os.path.join(base_path, '2_gwangmyung_signal.csv'), encoding='cp949')
sign = pd.read_csv(os.path.join(base_path, '2_gwangmyung_sign.csv'), encoding='cp949')
fence = pd.read_csv(os.path.join(base_path, '2_gwangmyung_fence.csv'), encoding='cp949')
red_surface = pd.read_csv(os.path.join(base_path, '2_gwangmyung_red_surface.csv'), encoding='cp949')
traffic_camera = pd.read_csv(os.path.join(base_path, '2_gwangmyung_traffic_camera.csv'), encoding='cp949')
stop = pd.read_csv(os.path.join(base_path, '2_gwangmyung_stop.csv'), encoding='cp949')

facilities = pd.concat([
    cctv,crosswalk,yellow_carpet,signal,sign,fence,red_surface,traffic_camera,stop],ignore_index=True)
print(facilities['카테고리'].value_counts())
facilities.head(5)
```

```python
# 스쿨존별로 300m 이내 시설물 개수 카운트

RADIUS = 300  # 미터
categories = facilities['카테고리'].unique()
school_coordinates = school_df[['대상시설명','위도','경도']].drop_duplicates()

results = []

for _, school in school_coordinates.iterrows():
    # 해당 학교와 모든 시설물 간 거리 벡터 계산
    distances = haversine_m(
        school['위도'], school['경도'],
        facilities['위도'].values,
        facilities['경도'].values
    )
    
    # 300m 이내 시설물만 필터
    nearby = facilities[distances <= RADIUS]
    
    # 카테고리별 카운트
    counts = nearby['카테고리'].value_counts()
    
    row = {'시설물명': school['대상시설명'], '위도': school['위도'], '경도': school['경도']}
    for cat in categories:
        row[cat] = counts.get(cat, 0)
    
    results.append(row)

result_df = pd.DataFrame(results)
result_df
```
